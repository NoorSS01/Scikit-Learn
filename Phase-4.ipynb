## Unsupervised Learning with Scikit-learn


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, load_wine
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

KMeans Clustering


```python
X, _ = make_blobs(n_samples=350, centers=4, cluster_std=1.1, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)

print(f"Cluster centers:{kmeans.cluster_centers_}")
print(f"Inertia: {kmeans.inertia_:.2f}")
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=35)
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    c='red',
    marker='X',
    s=180,
)
plt.title('KMeans Clustering Result')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(alpha=0.2)
plt.show()
```

PCA Dimensionality Reduction


```python
wine = load_wine(as_frame=True)
X_wine = wine.data
y_wine = wine.target

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_wine)

print(f"Explained variance ratio: {pca.explained_variance_ratio_}")
print(f"Total explained variance: {pca.explained_variance_ratio_.sum():.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_wine, cmap='plasma', alpha=0.8)
plt.title('Wine Dataset after PCA (2 Components)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid(alpha=0.2)
plt.show()
```
